## 5. Box plot numeric features grouped by label classes (na vaulues filled by median of column)

```python
train=train.select_dtypes(exclude='object')
train.fillna(train.median(),inplace=True)
```

```python
for predictor in train.columns:
    for label in labels.columns:
        df=pd.DataFrame({predictor:train[predictor].values,label:labels[label].values})
        df.boxplot(by=label,rot=45,figsize=(10,5));
```
